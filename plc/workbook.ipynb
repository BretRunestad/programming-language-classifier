```python
def tuple_maker(adict):
    lista = []
    for key in adict:
        lista.append(key)
    tup = tuple(lista)
    return tup
```

```python
def code_sucker():
    codelist = []
    for subdir, dirs, files in os.walk("bench/"):
        for fname in files:
            if fname.endswith(tuple_maker(filename_dict)):
                #print(fname)
                with open(os.path.join(subdir, fname)) as current_file:
                    codelist.append(current_file.read()) 
    return codelist
```

```python
def length_getter(code):
    return len(code)
```

```python
def character_counter(code, char):
    counter = 0
    for _ in code:
        if _ == char:
            counter+=1
    return counter

def character_ratio(code, char):
    value = character_counter(code, char)/len(code)
    return value
    
```

```python
def comma_counter(code):
    counter = 0
    for _ in code:
        if _ == ',':
            counter+=1
    return counter

def semicolon_counter(code):
    counter = 0
    for _ in code:
        if _ == ';':
            counter+=1
    return counter
```

```python
def string_finder(string, code):
    value = len(re.findall(string, code))
    return value

def string_ratio(string, code):
    value = string_finder(string, code)/len(code)
    return value
```

```python
def data_frame_generator():
    codelist = code_sucker()
    typelist = type_getter()
    df = pd.DataFrame(typelist, index=range(386))
    df.columns = ["Language"]
    df["Code"] = codelist
    df['Language'] = df.Language.apply(lambda x:x.lower())
    for string in word_list:
        def sub_function(code):
            x = string_ratio(string, code)
            return x
        df[string] = df.Code.apply(sub_function)
    return df
```

```python
def carat(code):
    x =character_ratio(code, '^')
    return x

def comma(code):
    x =character_ratio(code, ',')
    return x

def semicolon(code):
    x =character_ratio(code, ';')
    return x

def ampersand(code):
    x =character_ratio(code, '&')
    return x

def exclamation(code):
    x =character_ratio(code, '!')
    return x

def pipe(code):
    x =character_ratio(code, '|')
    return x
```

```python
def let_ratio(code):
    return string_ratio('let', code)

def end_ratio(code):
    return string_ratio('end', code)

def fun_ratio(code):
    return string_ratio('fun', code)

def defn_ratio(code):
    return string_ratio('defn', code)

def def_ratio(code):
    return string_ratio('def', code)

def function_ratio(code):
    return string_ratio('function', code)

def return_ratio(code):
    return string_ratio('return', code)

def define_ratio(code):
    return string_ratio('define', code)

def doublecolon_ratio(code):
    return string_ratio('::', code)

def check_ratio(code):
    return string_ratio('check', code)

def make_ratio(code):
    return string_ratio('make', code)

def format_ratio(code):
    return string_ratio('.format', code)

def arrow_ratio(code):
    return string_ratio('->', code)

def dollar_ratio(code):
    return string_ratio('$', code)

def printf_ratio(code):
    return string_ratio('printf', code)
```
